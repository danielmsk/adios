```r
scientific_notation <- function(x){
    x <- format(x, scientific=TRUE)
#     x <- gsub("^(.*)e", "'\\1'e", x)
#     x <- gsub("e", "%*%10^", x)
#     x <- gsub('\\+', '', x)
    parse(text=x)
}

plot.lmfit <- function(dset, title, setidx, k){
    par(lwd=2,font.axis=2,font.lab=2)
    rst_lm <- glm(paste(cn[k],"~ Age"), data=dset, family="gaussian")
    cortest<-cor.test(dset$Age,dset[,k])
    sum_rst_lm <- summary(rst_lm)

    p_value <- sum_rst_lm$coefficients[8]
    eff_size <- sum_rst_lm$coefficients[2]
    
    plot(d$Age[setidx], d[setidx, k], main=paste(title,"\np=", scientific_notation(p_value), ", R2=", round(cortest$estimate,3) ), xlab="Age", ylab=cn[k])
    abline(rst_lm,col="royalblue")
    rst <- c(p_value, eff_size, cortest$estimate)
    return (rst)
}

# k <- 5
rst <- c()
for (k in protidx){
    jpeg(filename = paste("dist_protein_exp/",cn[k],".jpg",sep=""), width = 900, height = 300, units = "px",pointsize = 15, quality = 90, bg = "white")
    par(mfrow=c(1,3))
    rst_m <- plot.lmfit(d[male_idx,], "Male", male_idx, k)
    rst_f <- plot.lmfit(d[female_idx,], "Female", female_idx, k)
    rst_a <- plot.lmfit(d, "All", all_idx, k)
    rst <- rbind(rst, c(cn[k], rst_m[1], rst_m[2], rst_m[3], rst_f[1], rst_f[2], rst_f[3], rst_a[1], rst_a[2], rst_a[3]))
    dev.off()
}

colnames(rst) <- c('protein_name', 'p_male', 'beta_male', 'corr_male', 'p_female', 'beta_female','corr_female', 'p_all', 'beta_all', 'corr_all' ) 
write.table(rst, file="protein_exp_reg_p_beta.txt", sep="\t", row.names=FALSE, quote=FALSE, append=FALSE)
```

```r
svmfit_1st <- function(dset, title, setidx, k){
    # modelsvm <- svm(model = paste(cn[k],"~ Age"), data=tset)
    modelsvm <- svm(x = dset[,k], y=dset$Age)
    # modelsvm <- svm(x = dset$Age, y=dset[,k] )
    pred = predict(modelsvm, dset[,k])

    plot(dset[,k], dset$Age, xlab=cn[k], main=title, ylab="Age")
    points(dset[,k], pred, col="red", pch=18)
#     plot(dset$Age, dset[,k], xlab="Age", ylab=cn[k])
#     points(pred,dset[,k], col="red", pch=18)
    r2 <- R2(dset$Age, pred, form = "traditional")
    rmse <- RMSE(dset$Age, pred)
    mae <- MAE(dset$Age, pred)
    r1<-c(r2, rmse, mae)
    return (r1)
}

rst <- c()
for (k in protidx){
# for (k in c(10:20)){
    jpeg(filename = paste("dist_svm_1st/",cn[k],".jpg",sep=""), width = 900, height = 300, units = "px",pointsize = 15, quality = 90, bg = "white")
    par(mfrow=c(1,3))
    rst_m <- svmfit_1st(d[male_idx,], "Male", male_idx, k)
    rst_f <- svmfit_1st(d[female_idx,], "Female", female_idx, k)
    rst_a <- svmfit_1st(d, "All", all_idx, k)
    rst <- rbind(rst, c(cn[k], rst_m[1], rst_m[2], rst_m[3], rst_f[1], rst_f[2], rst_f[3], rst_a[1], rst_a[2], rst_a[3]))
    dev.off()
}


colnames(rst) <- c('protein_name', 'r2_male', 'rmse_male', 'mae_male', 'r2_female', 'rmse_female','mae_female', 'r2_all', 'rmse_all', 'mae_all' ) 
write.table(rst, file="protein_exp_svm_1st.txt", sep="\t", row.names=FALSE, quote=FALSE, append=FALSE)



```
