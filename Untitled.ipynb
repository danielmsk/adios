# JT test

```r
A <- as.matrix(c(99,114,116,127,146,111, 125,143,148,157,133,139, 149, 160, 184))
g <- c(rep(1,5),rep(2,5),rep(3,5))
tapply(A, g, median)
A
JT.test(data = A, class = g, labs = c("GRP 1", "GRP 2", "GRP 3"), alternative = "two-sided")
```
