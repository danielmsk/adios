```r
male_idx <- which(d$Sex=="Male")
female_idx <- which(d$Sex=="Female")

# jpeg(filename = paste("fig_popmaf_histogram.jpg",sep=""), width = 600, height = 600, units = "px",pointsize = 15, quality = 90, bg = "white")


cn = colnames(d)
k <- 9

dset <- dmale
title <- "Male"
setidx <- male_idx

plot.lmfit <- function(dset, title, setidx){
    rst_lm <- glm(paste(cn[k],"~ Age"), data=dset, family="gaussian")
    cortest<-cor.test(dset$Age,dset[,k])
    sum_rst_lm <- summary(rst_lm)

    p_value <- sum_rst_lm$coefficients[8]
    eff_size <- sum_rst_lm$coefficients[2]

    plot(d$Age[setidx], d[setidx, k], main=paste(title,"\n", p_value, ", R2=", round(cortest$estimate,3) ), xlab="Age", ylab=cn[k])
    abline(rst_lm,col="royalblue")
}

plot.lmfit(d[male_idx,], "Male", male_idx)
plot.lmfit(d[female_idx,], "Female", female_idx)
plot.lmfit(d, "All", male_idx)


plot(d$Age[female_idx], d$CGA.FSHB[female_idx], main="Female")
# dev.off()
```
